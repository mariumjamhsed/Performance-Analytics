```python
## Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## import data sets

ipl=pd.read_csv("data/ipl-matches.csv")
ipl_runs=pd.read_csv( "data/batsman_runs_ipl.csv")
batsman=pd.read_csv("data/batsman_season_record.csv")
```

```python
batsman
```

Which team won the most matches?

```python
ipl["WinningTeam"].value_counts(ascending=False)
```

Winning percentage per team

```python
ipl["WinningTeam"].value_counts(ascending=False)/len(ipl)
```

Average score of each player

```python
ipl_runs.groupby("batter")["batsman_run"].mean().sort_values(ascending=False)
```

# Bar chart of top 5 winning team

```python
ipl["WinningTeam"].value_counts(ascending=False).head(5).plot(kind="bar")
plt.title("Top 5 Winning Team")
```

```python
final=ipl[ipl["MatchNumber"]=="Final"]
```

```python
final["WinningTeam"].value_counts(ascending=False).sort_values(ascending=False).plot(x="batsman_run", kind="bar")
plt.title("Team with most final Appearances")
```

The batsman graph for 3 years

```python
batsman.plot(kind="bar",x="batsman")
plt.show()
```

Top venue for high scores

```python
ipl["Venue"].value_counts(ascending=False).sort_values(ascending=False).plot(kind="bar",figsize=(10,15))
```
